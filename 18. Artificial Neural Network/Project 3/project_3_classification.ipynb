```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
import tensorflow as tf 
Sequential=tf.keras.models.Sequential
Dense=tf.keras.layers.Dense
Dropout=tf.keras.layers.Dropout
```

```python
len(df.columns)
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test,y_test), epochs=600)
```

```python
losses=pd.DataFrame(model.history.history)
losses.plot() #Clearly seen that this model is overfitting
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
EarlyStopping=tf.keras.callbacks.EarlyStopping
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test,y_test), epochs=600,callbacks=[early_stop])
```

```python
losses=pd.DataFrame(model.history.history)
losses.plot()
```

```python
Dropout=tf.keras.layers.Dropout
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test,y_test), epochs=600,callbacks=[early_stop])
```

```python
losses=pd.DataFrame(model.history.history)
losses.plot()
```

```python
y_pred = model.predict(X_test)
predictions = np.round(y_pred).astype(int)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
